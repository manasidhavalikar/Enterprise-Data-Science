```python
import pandas as pd
import plotly.express as plt_ex
```

```python
def get_data():
  data_fetch = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
  data = pd.read_csv(data_fetch, delimiter=',')
  return data
```

```python
def infectors():
  
  data = get_data()

  dataThreeCountries = data[(data['location']=='India') | (data['location']=='Italy') | (data['location']=='France')]
  dataThreeCountries.loc[: , ('infection_rate')] = dataThreeCountries['total_cases']/dataThreeCountries['population']

  fig = plt_ex.line(dataThreeCountries, x='date', y='infection_rate', color='location', title='The relative cases overtime of Covid-19 infectors for India, Italy and France.', 
             labels={
                 'date':'COVID-19 spread till date',
                 'infection_rate': 'COVID-19 Infection Rates (absolute Covid cases/population size)',
                 'location': 'Countries'
             })
  fig.update_layout(font_family="Times New Roman", yaxis={'tickformat': ',.0%'}, paper_bgcolor="lightblue")
  fig.show()
  return 0
```

```python
infectors()
```

```python
def vaccination_rate():
  
  data = get_data()

  dataThreeCountries = data[(data['location']=='India') | (data['location']=='Italy') | (data['location']=='France')]
  dataThreeCountries.loc[: , ('vaccination_rate')] = dataThreeCountries['people_vaccinated']/dataThreeCountries['population']

  fig = plt_ex.line(dataThreeCountries, x='date', y='vaccination_rate', color='location', title='The vaccination rate for India, Italy and France',
             labels={
                 'date': 'COVID-19 spread till date',
                 'vaccination_rate': 'COVID-19 vaccination rate (percentage of the population) over time',
                 'location': 'Countries'
             })
  fig.update_layout(font_family="Times New Roman", yaxis={'tickformat': ',.0%'}, paper_bgcolor="beige")
  #fig.update_layout(paper_bgcolor="beige")
  fig.show()
  return 0
```

```python
vaccination_rate()
```
